```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

hdi_df = pd.read_csv('/content/drive/MyDrive/Dataset/Human_Development_Index_Dataset.csv',encoding='latin-1')
```

```python
# Task 1: Extract latest year

# Identify unique years in the dataset.
print(hdi_df['year'].unique())
```

```python
# Filter the dataset to include only observations from the year 2022.
hdi_df_2022 = hdi_df[hdi_df['year']==2022]

# Save the filtered dataframe as hdi 2022 df
hdi_2022_df = hdi_df_2022.copy()
```

```python
hdi_2022_df = hdi_2022_df.drop_duplicates()


num_cols = ['hdi', 'life_expectancy', 'gross_inc_percap']
for col in num_cols:
    hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')
```

```python
# Mean imputation (justified: minimal missing values)
hdi_2022_df[num_cols] = hdi_2022_df[num_cols].fillna(
    hdi_2022_df[num_cols].mean()
)
```

```python
# Task 4: Basic Statistics
hdi_2022_df['hdi'].mean()
hdi_2022_df['hdi'].median()
hdi_2022_df['hdi'].std()


hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax()][['country','hdi']]
hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin()][['country','hdi']]
```

```python
# Task 5: Filtering & Sorting
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]
high_hdi.sort_values('gross_inc_percap', ascending=False).head(10)
```

```python
# Task 6: HDI Category
def hdi_category(value):
    if value < 0.55:
        return 'Low'
    elif value < 0.70:
        return 'Medium'
    elif value < 0.80:
        return 'High'
    else:
        return 'Very High'

hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(hdi_category)

hdi_2022_df.to_csv('HDI_category_added.csv', index=False)
```

```python
# ==========================================================
# Problem 1B: HDI Visualization & Trend Analysis (2020â€“2022)
# ==========================================================


hdi_1b = hdi_df[hdi_df['year'].isin([2020, 2021, 2022])].copy()
hdi_1b.to_csv('HDI_problem1B.csv', index=False)


# Cleaning
hdi_1b = hdi_1b.drop_duplicates()
hdi_1b[['hdi']] = hdi_1b[['hdi']].apply(pd.to_numeric, errors='coerce')
hdi_1b = hdi_1b.dropna(subset=['hdi','country','year'])
```

```python
# ==========================================================
# Problem 2: Advanced HDI Exploration â€“ South Asia
# ==========================================================


south_asia = ['Afghanistan','Bangladesh','Bhutan','India','Maldives','Nepal','Pakistan','Sri Lanka']
hdi_sa = hdi_df[hdi_df['country'].isin(south_asia)].copy()
hdi_sa.to_csv('HDI_SouthAsia.csv', index=False)
hdi_sa.head()
```

```python
# Composite Score
hdi_sa['Composite Score'] = (0.30 * hdi_sa['life_expectancy'] + 0.30 * hdi_sa['gross_inc_percap'])

hdi_sa_cs = hdi_sa.dropna(subset=['Composite Score'])
avg_composite = (hdi_sa_cs.groupby("country", as_index=False)["Composite Score"].mean())
ranked = avg_composite.sort_values("Composite Score", ascending=False).head(5)
ranked
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=ranked, x='Composite Score', y='country')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.show()

```

```python
# Outlier Detection (IQR)
def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (column < lower) | (column > upper)

# Detect outliers
hdi_sa['HDI Outlier'] = detect_outliers(hdi_sa['hdi'])
hdi_sa['GNI Outlier'] = detect_outliers(hdi_sa['gross_inc_percap'])


#scatter plot of GNI per Capita vs HDI
plt.figure(figsize=(7,5))

plt.scatter(hdi_sa['gross_inc_percap'],hdi_sa['hdi'],label='Normal')

outliers = hdi_sa[hdi_sa['HDI Outlier'] | hdi_sa['GNI Outlier']]

plt.scatter(outliers['gross_inc_percap'],outliers['hdi'],label='Outliers')

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outliers in South Asia: HDI vs GNI (2022)")
plt.legend()
plt.show()
```

```python
# Correlation Analysis
print('Life Expectancy vs HDI:', hdi_sa['life_expectancy'].corr(hdi_sa['hdi']))
print('GNI vs HDI:', hdi_sa['gross_inc_percap'].corr(hdi_sa['hdi']))
```

```python
# Gap Analysis
hdi_sa['GNI_HDI_Gap'] = hdi_sa['gross_inc_percap'] - hdi_sa['hdi']

hdi_sa.sort_values('GNI_HDI_Gap', ascending=False).head(3)
hdi_sa.sort_values('GNI_HDI_Gap').head(3)
```

```python
# ==========================================================
# Problem 3: South Asia vs Middle East (2020â€“2022)
# ==========================================================

middle_east = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
"Kuwait", "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia", "Syria",
"United Arab Emirates", "Yemen"]

sa_2020_22 = hdi_1b[hdi_1b['country'].isin(south_asia)]
me_2020_22 = hdi_1b[hdi_1b['country'].isin(middle_east)]


sa_2020_22.to_csv('HDI_SouthAsia_2020_2022.csv', index=False)
me_2020_22.to_csv('HDI_MiddleEast_2020_2022.csv', index=False)
```

```python
# Descriptive Statistics
print('South Asia Mean & Std:', sa_2020_22['hdi'].mean(), sa_2020_22['hdi'].std())
print('Middle East Mean & Std:', me_2020_22['hdi'].mean(), me_2020_22['hdi'].std())
```

```python
# Top and Bottom Performers:

# Compute mean HDI per country
mean_hdi_sa = sa_2020_22.groupby("country")["hdi"].mean()
mean_hdi_me = me_2020_22.groupby("country")["hdi"].mean()

# Top 3 performers
top3_sa = mean_hdi_sa.sort_values(ascending=False).head(3).reset_index()
top3_me = mean_hdi_me.sort_values(ascending=False).head(3).reset_index()

# Bottom 3 performers
bottom3_sa = mean_hdi_sa.sort_values(ascending=True).head(3).reset_index()
bottom3_me = mean_hdi_me.sort_values(ascending=True).head(3).reset_index()

print("South Asia â€“ Top 3:\n", top3_sa)
print("\nSouth Asia â€“ Bottom 3:\n", bottom3_sa)

print("\nMiddle East â€“ Top 3:\n", top3_sa)
print("\nMiddle East â€“ Bottom 3:\n", bottom3_sa)
```

```python
# Add labels
top3_sa['category'] = 'Top 3'
bottom3_sa['category'] = 'Bottom 3'
top3_me['category'] = 'Top 3'
bottom3_me['category'] = 'Bottom 3'

top3_sa['region'] = 'South Asia'
bottom3_sa['region'] = 'South Asia'
top3_me['region'] = 'Middle East'
bottom3_me['region'] = 'Middle East'

# Combine ONLY for plotting
plot_df = pd.concat([
    top3_sa, bottom3_sa,
    top3_me, bottom3_me
])


plt.figure(figsize=(12,6))

sns.barplot(
    data=plot_df,
    x='category',      # Top 3 vs Bottom 3
    y='hdi',
    hue='region'       # Side-by-side bars
)

plt.xlabel('Performer Category')
plt.ylabel('Mean HDI (2020â€“2022)')
plt.title('Comparison of Top and Bottom HDI Performers by Region')
plt.legend(title='Region')
plt.show()

```

```python
# Metric Comparisons:
metrics = ["gender_development", "life_expectancy"]

# Plot each metric separately
for metric in metrics:
    plt.figure(figsize=(8,5))
    values = [sa_2020_22[metric].mean(), me_2020_22[metric].mean()]
    plt.bar(["South Asia", "Middle East"], values, color=["cyan", "maroon"])
    plt.ylabel(metric.replace("_", " ").title())
    plt.title(f"{metric.replace('_', ' ').title()} Comparison: South Asia vs Middle East (2020â€“2022)")
    for i, v in enumerate(values):
        plt.text(i, v + 0.02*max(values), f"{v:.2f}", ha='center', fontweight='bold')
    plt.show()
```

```python
# HDI Disparity:

# Compute HDI range (max - min)
range_sa = sa_2020_22["hdi"].max() - sa_2020_22["hdi"].min()
range_me = me_2020_22["hdi"].max() - me_2020_22["hdi"].min()

# Compute coefficient of variation (CV = std / mean)
cv_sa = sa_2020_22["hdi"].std() / sa_2020_22["hdi"].mean()
cv_me = me_2020_22["hdi"].std() / me_2020_22["hdi"].mean()

print(f"South Asia HDI Range: {range_sa:.3f}")
print(f"Middle East HDI Range: {range_me:.3f}")
print(f"South Asia HDI CV: {cv_sa:.3f}")
print(f"Middle East HDI CV: {cv_me:.3f}")

# Identify which region exhibits more variation
if cv_sa > cv_me:
    print("South Asia exhibits more relative variation in HDI.")
else:
    print("Middle East exhibits more relative variation in HDI.")
```

```python
# Correlation Analysis:

# Metrics to correlate with HDI
metrics = ["gender_development", "life_expectancy"]

# Function to compute correlations and plot scatter with trendline
def correlation_analysis(df, region_name):
    print(f"\n--- {region_name} ---")
    for metric in metrics:
        corr = df["hdi"].corr(df[metric])
        print(f"Correlation between HDI and {metric.replace('_',' ').title()}: {corr:.3f}")

        # Scatter plot with trendline
        plt.figure()
        sns.scatterplot(x=metric, y="hdi", data=df)
        plt.xlabel(metric.replace("_"," ").title())
        plt.ylabel("HDI")
        plt.title(f"{region_name}: HDI vs {metric.replace('_',' ').title()}")
        plt.show()

# Run analysis for both regions
correlation_analysis(sa_2020_22, "South Asia")
correlation_analysis(me_2020_22, "Middle East")
```

```python
# Outlier Detection:

# Function to detect outliers using 1.5*IQR
def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (column < lower) | (column > upper)

# Detect outliers for HDI and GNI per Capita
sa_2020_22['hdi_outlier'] = detect_outliers(sa_2020_22['hdi'])
sa_2020_22['gni_outlier'] = detect_outliers(sa_2020_22['gross_inc_percap'])

me_2020_22['hdi_outlier'] = detect_outliers(me_2020_22['hdi'])
me_2020_22['gni_outlier'] = detect_outliers(me_2020_22['gross_inc_percap'])


# outliers_sa_hdi = detect_outliers(sa_2020_22, "hdi")
# outliers_sa_gni = detect_outliers(sa_2020_22, "gross_inc_percap")

# outliers_me_hdi = detect_outliers(me_2020_22, "hdi")
# outliers_me_gni = detect_outliers(me_2020_22, "gross_inc_percap")

# Scatter plot function
plt.figure()
plt.scatter(
    sa_2020_22['gross_inc_percap'],
    sa_2020_22['hdi']
)
plt.scatter(
    sa_2020_22.loc[sa_2020_22['hdi_outlier'] | sa_2020_22['gni_outlier'], 'gross_inc_percap'],
    sa_2020_22.loc[sa_2020_22['hdi_outlier'] | sa_2020_22['gni_outlier'], 'hdi']
)

plt.title("South Asia: HDI vs GNI per Capita (Outliers Highlighted)")
plt.xlabel("GNI per Capita Index")
plt.ylabel("HDI")
# plt.tight_layout()
plt.legend()
plt.show()
```

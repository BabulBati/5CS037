```python
# Problem 3
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

print("Add\n", x+v)
print("Subtract\n",y-w)
print("Multiply\n",x*5)
print("Square\n",y**2)
print("Dot v.w\n",np.dot(v,w))
print("Dot x.v\n",np.dot(x,v))
print("Dot x.y\n",np.dot(x,y))
print("Concatenate row\n",np.concatenate((x,y), axis=0))
print("Concatenate column\n",np.vstack((v,w)))
```

```python
# Problem 4
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# Prove A.A−1 = I
A_inv = np.linalg.inv(A)
print(A @ A_inv)
```

```python
# Prove AB ̸= BA.
print("AB =\n", A @ B)
print("BA =\n", B @ A)
```

```python
# Prove (AB)T = BTAT
print((A@B).T)
print((B.T)@(A.T))
```

```python
A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])

B = np.array([-1,-3,9])
X = np.linalg.inv(A) @ B
print(X)
```

```python
M1 = [[1]*1000 for _ in range(1000)]
M2 = [[1]*1000 for _ in range(1000)]

start = time.time()
Mout = [[sum(a*b for a,b in zip(row,col)) for col in zip(*M2)] for row in M1]
end = time.time()

print("Python matrix multiplication time:", end - start)
```
